```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
resolution = 72.27
textwidth = 307.28987/resolution
textheight = .85*261.14662/resolution
textiny, texsmall, texnormal, texlarge = 5.5, 7., 8., 9.
colors = ('#33b5e5','#99cc00','#ff4444','#aa66cc','#ffbb33')

plt.rcParams.update({
    #'font.family': 'CMU Sans Serif',
    'font.family': 'sans-serif',
    'font.sans-serif': ['CMU Sans Serif'],
    'mathtext.fontset': 'custom',
    'mathtext.rm': 'sans',
    'mathtext.it': 'sans:italic',
    'figure.facecolor': '1.0',
    'font.size': texsmall,
    'axes.titlesize': texsmall,
    'axes.labelsize': texsmall,
    'legend.fontsize': texsmall,
    'xtick.labelsize': textiny,
    'ytick.labelsize': textiny,
    'axes.linewidth': .4,
    'xtick.major.size': 1.7,
    'ytick.major.size': 1.7,
    'xtick.minor.size': 1.2,
    'ytick.minor.size': 1.2,
    'axes.color_cycle': colors,
    'figure.figsize': [textwidth,textheight],
})
```

```python
import os

path = os.path.expanduser('~') + '/qcd/ebe/results/'
thevn = 2,3,4
```

```python
atlas = {n: {'cent': [], 'vnrp': [], 'deltavn': []} for n in thevn}
with open(path + 'exp/atlas/fits/flowparams-5.dat') as f:
    for l in f:
        vn, cent, vnrp, deltavn, *rest = l.split()
        vn = int(vn)
        cent = 0.5*sum(map(int,cent.split('-')))
        vnrp = float(vnrp)
        deltavn = float(deltavn)
        atlas[vn]['cent'].append(cent)
        atlas[vn]['vnrp'].append(vnrp)
        atlas[vn]['deltavn'].append(deltavn)
```

```python
from glob import glob

remove_outlier = lambda x: np.delete(x, 3, axis=0)

model = {n: {'cent': [], 'vnrp': [], 'deltavn': []} for n in thevn}

for vn in thevn:
    _vnrp = []
    _deltavn = []
    for fn in sorted(glob(path + 'fits/flowparams_glb_v{}_*.dat'.format(vn))):
        vnrp, deltavn = remove_outlier(np.loadtxt(fn, usecols=(0,1))).T
        cent = 0.5*sum(map(int,fn.rstrip('.dat')[-5:].split('-')))
        model[vn]['cent'].append(cent)
        _vnrp.append(vnrp)
        _deltavn.append(deltavn)
    model[vn]['vnrp'] = np.array(_vnrp).T
    model[vn]['deltavn'] = np.array(_deltavn).T
```

```python
design = remove_outlier(np.loadtxt(path + 'design/lhs-design-glb-256.dat'))
X = np.array([[c] + x.tolist() for x in design for c in model[2]['cent']])
Xmin, Xmax = X.min(axis=0), X.max(axis=0)
Xrange = Xmax - Xmin
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, sharex='col', sharey='row', figsize=(textwidth,textheight))

handles = [plt.Line2D((0,1), (0,0), lw=2, color=colors[i]) for i in range(3)]
labels = [r'$0.{} < \eta/s < 0.{}$'.format(i, i+1) for i in range(3)]

for row, var, label in zip(axes, ('vnrp', 'deltavn'), (r'$v_n^\mathrm{RP}$', r'$\delta v_n$')):
    for ax, vn in zip(row, thevn):
        for i, x in enumerate(model[vn][var]):
            etas = design[i,3]
            cidx = int(etas/0.1)
            ax.plot(model[vn]['cent'], x, color=colors[cidx], lw=.3, alpha=.3)
            
        hatlas = ax.plot(atlas[vn]['cent'], atlas[vn][var], 'k-o', ms=2.25, lw=.8)
        
        ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(2))
        ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(2))
        
        ax.set_ylim(bottom=0)
        ax.set_xlim(0,69.5)

        if ax.is_first_row():
            ax.set_title('$v_{}$'.format(vn))
            ax.set_ylim(top=.13)
        else:
            ax.set_xlabel('Centrality %')
            ax.set_ylim(top=.06)
            
        if ax.is_first_col():
            ax.set_ylabel(label)
            
        if ax.is_last_col() and ax.is_first_row():
            ax.legend(hatlas + handles, ['ATLAS'] + labels)
            
plt.tight_layout(pad=0, h_pad=0, w_pad=0)
plt.savefig('bgparams.pdf')
```

```python
from sklearn.gaussian_process import GaussianProcess
import warnings
```

```python
# run 'ipcluster start' before executing this cell

from IPython.parallel import Client
import itertools

def train(vn, var, Ntrain=255*6):
    from sklearn.gaussian_process import GaussianProcess
    import warnings
    
    gp = GaussianProcess(
                 theta0 = 0.2*Xrange,
                 thetaL = .01*Xrange,
                 thetaU = 5*Xrange,
                 nugget = .01**2,
                 random_start = 1,
                 regr = 'linear'
                 )
    
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        gp.fit(X[-Ntrain:], model[vn][var].ravel()[-Ntrain:])
        
    return vn, var, gp

rc = Client()
dview = rc[:]
dview['X'] = X
dview['Xrange'] = Xrange
dview['model'] = model

allvn, allvar = zip(*itertools.product(thevn, ('vnrp','deltavn')))
res = dview.map_sync(train, allvn, allvar)
GPs = {n: {} for n in thevn}
for vn, var, gp in res:
    GPs[vn][var] = gp
```

```python
from sklearn.externals import joblib
joblib.dump(GPs, '/home/jonah/xtmp/gp', compress=3)
```

```python
def emuplot(fname, *args, cents=model[2]['cent']):
    args = list(args)
    
    fig, axes = plt.subplots(nrows=2, ncols=3, sharex='col', sharey='row', figsize=(textwidth,textheight))
    
    for row, var, label in zip(axes, ('vnrp', 'deltavn'), (r'$v_n^\mathrm{RP}$', r'$\delta v_n$')):
        for ax, vn in zip(row, thevn):
            ax.plot(atlas[vn]['cent'], atlas[vn][var], 'k-o', ms=2.25, lw=.8, label='ATLAS')
            
            gp = GPs[vn][var]
            mean, mse = gp.predict([[c] + args for c in cents], eval_MSE=True)
            err = np.sqrt(mse)
            ax.fill_between(cents, mean+err, mean-err, color=colors[0], alpha=.2)
            ax.plot(cents, mean, label=r'Emulator $\pm1\sigma$')
        
            ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(2))
            ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(2))
        
            ax.set_ylim(bottom=0)
            ax.set_xlim(0,69.5)
    
            if ax.is_first_row():
                ax.set_title('$v_{}$'.format(vn))
                ax.set_ylim(top=.12)
            else:
                ax.set_xlabel('Centrality %')
                ax.set_ylim(top=.06)
                
            if ax.is_first_col():
                ax.set_ylabel(label)
                
            if ax.is_first_row() and not (ax.is_first_col() or ax.is_last_col()):
                info = '\n'.join(
                                 '{} = {}'.format(a,b) for a,b in zip(
                                    ('Norm.', r'$\alpha$', r'$\tau_0$', r'$\eta/s$', r'$\tau_\Pi$'), args)
                                )
                ax.annotate('Glauber\n' + info, (.1,.55), xycoords='axes fraction', ha='left', va='center', size='x-small')
                
            if ax.is_last_col() and ax.is_first_row():
                ax.legend()
        
            
    plt.tight_layout(pad=0, h_pad=0, w_pad=0)
    plt.savefig(fname + '.pdf')
```

```python
emuplot('emu0', 55., .12, .6, .08, .5)
emuplot('emu1', 55., .12, .6, .04, .5)
emuplot('emu2', 30., .2, .5, .04, .5)
```
